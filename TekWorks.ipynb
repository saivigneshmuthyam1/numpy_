Indexing and Slicing:
You are working as a Data Analyst at a fruit distribution company.

The company stores the number of fruit boxes shipped per day (for 1 week) in a 2D NumPy array where:
Each row represents a type of fruit.
Each column represents a day of the week (Mon–Sun).


Your Tasks
Create the 2D NumPy array

Create an array named shipments using the above data.
Access single elements (Indexing):
Get the number of Apples shipped on Wednesday.
Get the number of Oranges shipped on Sunday.
Access multiple elements (Slicing):
Extract all the weekday shipments (Mon–Fri) for all fruits.
Extract only the weekend shipments (Sat & Sun).
Extract the shipments of Apples from Wednesday to Saturday.
Bonus (Practice thinking):
Replace the Saturday shipment of Apples (currently 80) with 100 using indexing.


```python
import numpy as np

# Data: Number of fruit boxes shipped per day for 1 week
# Rows: Fruit types (Apples, Oranges, Bananas)
# Columns: Days of the week (Mon-Sun)
data = [
    [100, 120, 110, 130, 150, 80, 90],  # Apples
    [80, 90, 85, 95, 100, 70, 75],      # Oranges
    [110, 100, 105, 115, 120, 90, 85]   # Bananas
]

# Create the 2D NumPy array
shipments = np.array(data)

print("Shipments Array:")
print(shipments)
print("-" * 30)

# Access single elements (Indexing):
# Get the number of Apples shipped on Wednesday (Apples are row 0, Wednesday is column 2)
apples_wednesday = shipments[0, 2]
print(f"Apples shipped on Wednesday: {apples_wednesday}")

# Get the number of Oranges shipped on Sunday (Oranges are row 1, Sunday is column 6)
oranges_sunday = shipments[1, 6]
print(f"Oranges shipped on Sunday: {oranges_sunday}")
print("-" * 30)

# Access multiple elements (Slicing):
# Extract all the weekday shipments (Mon-Fri) for all fruits (all rows, columns 0 to 4)
weekday_shipments = shipments[:, 0:5]
print("Weekday Shipments:")
print(weekday_shipments)

# Extract only the weekend shipments (Sat & Sun) (all rows, columns 5 and 6)
weekend_shipments = shipments[:, 5:7]
print("\nWeekend Shipments:")
print(weekend_shipments)

# Extract the shipments of Apples from Wednesday to Saturday (Apples are row 0, columns 2 to 5)
apples_wed_to_sat = shipments[0, 2:6]
print(f"\nApples shipments from Wednesday to Saturday: {apples_wed_to_sat}")
print("-" * 30)

# Bonus (Practice thinking):
# Replace the Saturday shipment of Apples (currently 80) with 100 using indexing
# Saturday shipment of Apples is at row 0, column 5
shipments[0, 5] = 100
print("Updated Shipments Array (Saturday Apples replaced):")
print(shipments)
```

```python
import numpy as np

# Array Creation
# Create a 1D array of numbers from 10 to 50.
arr_1d = np.arange(10, 51)
print("1D array from 10 to 50:", arr_1d)

# Create a 3x3 array of random numbers between 0 and 1.
arr_rand = np.random.rand(3, 3)
print("\n3x3 array of random numbers:\n", arr_rand)

# Array Attributes
# Find the shape, size, and data type of the array you created in the previous task.
print("\nAttributes of the random array:")
print("Shape:", arr_rand.shape)
print("Size:", arr_rand.size)
print("Data type:", arr_rand.dtype)

# Basic Operations
# Add 5 to every element of a 1D array.
arr_1d_plus_5 = arr_1d + 5
print("\n1D array with 5 added to each element:", arr_1d_plus_5)

# Multiply two 2D arrays element-wise.
arr_mult1 = np.array([[1, 2], [3, 4]])
arr_mult2 = np.array([[5, 6], [7, 8]])
arr_element_wise_mult = arr_mult1 * arr_mult2
print("\nElement-wise multiplication of two 2D arrays:\n", arr_element_wise_mult)

# Indexing and Slicing
# Create a 5x5 array of integers from 1 to 25.
arr_5x5 = np.arange(1, 26).reshape(5, 5)
print("\n5x5 array of integers from 1 to 25:\n", arr_5x5)

# Extract the first row.
first_row = arr_5x5[0, :]
print("\nFirst row of the 5x5 array:", first_row)

# Extract the last column.
last_column = arr_5x5[:, -1]
print("Last column of the 5x5 array:", last_column)

# Extract a subarray of the first three rows and first two columns.
subarray = arr_5x5[0:3, 0:2]
print("Subarray of the first three rows and first two columns:\n", subarray)

# Aggregation
# Create a 10-element array of random integers between 10 and 100.
arr_agg = np.random.randint(10, 101, 10)
print("\n10-element array of random integers:", arr_agg)

# Find the maximum value.
max_value = np.max(arr_agg)
print("Maximum value:", max_value)

# Find the minimum value.
min_value = np.min(arr_agg)
print("Minimum value:", min_value)

# Find the mean and standard deviation.
mean_value = np.mean(arr_agg)
std_dev = np.std(arr_agg)
print("Mean:", mean_value)
print("Standard deviation:", std_dev)

# Reshaping
# Convert a 1D array of numbers from 1 to 12 into a 3x4 matrix.
arr_1_to_12 = np.arange(1, 13)
arr_reshaped = arr_1_to_12.reshape(3, 4)
print("\n1D array reshaped into a 3x4 matrix:\n", arr_reshaped)

# Broadcasting
# Create a 3x3 matrix of ones. Add a 1D array [1, 2, 3] to each row.
arr_ones = np.ones((3, 3))
arr_to_add = np.array([1, 2, 3])
arr_broadcasted = arr_ones + arr_to_add
print("\n3x3 matrix of ones with [1, 2, 3] added to each row (broadcasting):\n", arr_broadcasted)

# Filtering
# Create a 1D array of random integers between 1 and 50.
arr_filter = np.random.randint(1, 51, 20)
print("\nOriginal array for filtering:", arr_filter)

# Extract all elements that are divisible by 5.
filtered_arr = arr_filter[arr_filter % 5 == 0]
print("Elements divisible by 5:", filtered_arr)

# Matrix Multiplication
# Create two 2x2 matrices and perform matrix multiplication.
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
matrix_mult_result = np.dot(matrix1, matrix2)
print("\nMatrix multiplication of two 2x2 matrices:\n", matrix_mult_result)

# Advanced Indexing
# Given a 4x4 array, use fancy indexing to extract the diagonal elements.
arr_4x4 = np.arange(1, 17).reshape(4, 4)
print("\nOriginal 4x4 array:\n", arr_4x4)

# Extract diagonal elements using fancy indexing
diagonal_elements = arr_4x4[[0, 1, 2, 3], [0, 1, 2, 3]]
print("Diagonal elements using fancy indexing:", diagonal_elements)
```
